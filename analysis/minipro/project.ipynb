```python
top_five_i = entry_Y_cty.loc[(entry_Y_cty['국적']=='일본') | (entry_Y_cty['국적']=='중국') | (entry_Y_cty['국적']=='미국')|
               (entry_Y_cty['국적']=='대만') | (entry_Y_cty['국적']=='홍콩') ]
top_five_i.loc[top_five_i['목적']=='소 계']
#'소 계'행 삭제
top_five = top_five_i[top_five_i['목적'] != '소 계']
```
