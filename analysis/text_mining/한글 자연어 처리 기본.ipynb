### 형태소 분석기는 트위터를 사용
- 워드클라우드를 사용하여 지문의 중요 단어들 표시
- 이미지(해당 내용을 좀더 부각)위에 마스킹 처리하여서 시각화
- stopword 항목을 통해서 부적절한 용어나, 제거한 단어들을 빼고 처리
- 연관어 유사어등등 항목들을 구성하여 표현

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import numpy as np
from PIL import Image
```

```python
with open('./data/alice.txt') as f:
    text = f.read()
text[:100], len(text)
```

```python
# 이미지 마스킹
alice_mask = np.array( Image.open( './data/alice_mask.png' ) )
alice_mask.shape
```

```python
# 시각화 준비
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 한글 처리
from korea_font import initKoreaFontLoad
initKoreaFontLoad()
```

```python
# 이미지 그리기
plt.figure(figsize=(8,8))
# interpolation='bilinear' <= 이미지를 그리는 처리에 중요한 부분
# 이중선형 보간법 설정, 1차원 선형보간법을 2차원으로 확장한다는 의미
plt.imshow( alice_mask, cmap=plt.cm.gray, interpolation='bilinear' )
# 축을 제거
plt.axis('off')
plt.show()
```

```python
STOPWORDS
```

```python
# 카운팅에 배제되는 단어
# 단어로 사용되지 않는 영어문자들을 모아둔것 : STOPWORDS
# 추가로 배제 가능
stopwords = set( STOPWORDS )
# 만약에 특정 단어를 추가하고 싶다면
stopwords.add( 'said' )
len(stopwords)
```

```python
# 워드 클라우드 생성 
# 워드 카운트 기능 제공
wc = WordCloud( background_color='white', 
           max_words=2000, 
           mask=alice_mask, 
           stopwords=stopwords)
```

```python
len( wc.words_ ), type(wc.words_)
```

### KoNLPy를 이용한 처리
- 기본으로 제공하는 내장 법안을 기준으로 분석
- 법률 문서 분석

```python
import nltk
```

```python
from konlpy.corpus import kobill
```

```python
kobill.fileids()
```

```python
kobill.fileids()[-1]
```

```python
doc_ko = kobill.open( kobill.fileids()[-1]).read()
```

```python
doc_ko[:100], len(doc_ko)
```

```python
from konlpy.tag import Okt
t = Okt()
```

```python
# 명사분석
tokens_ko = t.nouns( doc_ko )
tokens_ko[:10], len( tokens_ko )
```

```python
# 분석의 제목  설정
ko = nltk.Text( tokens_ko, name='결혼중개업의 관리에 관한 법률 일부개정법률안' )
```

```python
len(ko.tokens), len( set(ko.tokens) )
```

```python
# 시긱화
plt.figure(figsize=(12,6))
# 상위 30개만 노출, 빈도수를 표현
ko.plot( 30 )
plt.show()
```

```python
# 임의로 stopword를 설정
#stop_words = []

# 한단어는 제거시켜라
ko  = list( filter( lambda x: len(x)>1, ko) )
# 실제 의미를 가진 한글자도 같이 제거가 되므로, 조심
```

```python
# 필터 조건은 상황에 맞게 구성하여 배제
# 상황이 더 복잡하면 => 정규식으로 이동
stop_words = [ '.','(',')','%','-','의','을','다','만','로','제','호','및','이','항' ]
def myFileter( x ):
    if len(x)>1: # 한글자 제거
        return True
    else:
        if x not in stop_words: # 배제 단어에 없으면 회생
            return True
        else: # 배제 단어에 포함되어 있으면 제거
            return False

def myFileter2( x ):
    # 한글자 이상이거나, 배제 단어에 없으면 포함
    return  (len(x)>1 or x not in stop_words)
        
    
```

```python
ko  = list( filter( myFileter, ko) )
```

```python
len(ko), ko
```

```python
stop_words = [ '.','(',')','%','-','의','을','다','만','로','제','호','및','이','항' ]
ko = [ each_word  for each_word in ko if each_word not in stop_words  ] 
ko
```

```python
ko = nltk.Text( ko, name='하도급거래 공정화에 관한 법률 일부개정법률안' )
plt.figure(figsize=(12,6))
ko.plot( 30 )
plt.show()
```

```python
# 특정 단어의 빈도형태와 문장안의 위치를 파악
# 논설위원이나 작가들의 총 데이터를 분석해서 통계적을호 살펴보면, 
# 글의 특성을 예측해 낼수 있다 -> 글의 지문
plt.figure( figsize=(12, 6) )
ko.dispersion_plot([ '결혼','신상','위반' ])
```

```python
# 특정 단어의 주변 단어 => 텍스트 주변에 자주 나오는 표현
ko.concordance('결혼')
```

```python
# 연어 , 비슷한 의미를 내포하겟금 사용된 단어
ko.collocations()
```
