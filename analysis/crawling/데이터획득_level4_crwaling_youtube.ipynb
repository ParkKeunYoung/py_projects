```python
from selenium import webdriver as wd
```

```python
import urllib
```

```python
# 업로드순 정렬, 오늘( 현재시간 24시간이내 게시물)
target_url = 'https://www.youtube.com/results?search_query={}&sp=EgQIAhAB'
```

```python
keyword = '갤럭시노트10'
```

```python
target_url = target_url.format(urllib.parse.quote(keyword))
target_url
```

```python
# 웹드라이버 가동
driver = wd.Chrome('../tool/chromedriver_75.0.3770.140.exe')
```

```python
# 사이트 접속
driver.get(target_url)
```

```python
# 스크롤을 쳐서 데이터를 다 로드 시킨다.
# 데이터량이 엄청 많지 않을 때 가능
# ajax를 사용할 경우 서버가 반응해서 응답이오고 화면구성할 때가지 시간이 걸린다.
import time
for n in range(10):
    driver.execute_script('window.scrollBy(0,1000)')
    # 다음 작업을 코드 적으로 강제 대기 시킨 것
    # 데이터 추출 시 selenium 에서 지원하는 wait 기법을 사용하면 좋다. 
    time.sleep(1)
```

```python
# 영상 데이터 한줄 :
# ytd-video-renderer
# 영상 메다 정보 :
# #video-title
```

```python
videos = driver.find_elements_by_tag_name('ytd-video-renderer') # elements 는 다 찾는거, element는 하나만
len(videos),type(videos) # 오늘 올라온 영상의 갯 수 , why? 필터를 오늘로 설정하고 url을 긁어 왔기 때문에.
```

```python
for video in videos:
    # 영상의 타이틀
    print(video.find_element_by_id('video-title').get_attribute('title'))  # #video-title 쓸 필요 x #은 셀렉터만을 지칭하므로
    # 썸네일 thumbnail
    print(video.find_element_by_tag_name('img').get_attribute('src'))  # by_tag_name 으로 img(이미지) 가능.
    # 링크 
    print(video.find_element_by_id('video-title').get_attribute('href'))
    # 작성자
    print(video.find_element_by_css_selector('#byline>a').text)  # byline>:a a는 byline의 직계자식
    # 조회수 : metadata-line, # 작성시간
    for span in video.find_elements_by_css_selector('#metadata-line>span'):
        print(span.text)
        
        
```

```python
#     구조로 만들어보기     
    '''
    [
    { 'title':'xx', 'thum':'xx', 'link':'xx', 'author':'xx', 'views':'x', 'time':'' }
        ,{},{}, ......
    ]
    '''
```

```python
results = []
for video in videos:
    dic = {
    'title':  video.find_element_by_id('video-title').get_attribute('title')  # #video-title 쓸 필요 x #은 셀렉터만을 지칭하므로
    ,'thum':  video.find_element_by_tag_name('img').get_attribute('src')  # by_tag_name 으로 img(이미지) 가능.
    ,'link':  video.find_element_by_id('video-title').get_attribute('href')
    ,'author':video.find_element_by_css_selector('#byline>a').text  # byline>:a a는 byline의 직계자식
    ,'views' :video.find_elements_by_css_selector('#metadata-line>span')[0].text
    ,'time' :video.find_elements_by_css_selector('#metadata-line>span')[1].text
    }
    results.append( dic )
results
```

```python
for link in results[:2]:
    print(link['link'])
    # 해당 콘텐츠에 진입
    driver.get(link['link'])
    # 잠시대기 --> 반드시 신규페이지에서는 대기
    print('5초 대기시도')
    driver.implicitly_wait(5)  
    # 화면 스크롤
    for n in range(3):
        print('화면스크롤 시도')
        driver.execute_script('window.scrollBy(0,1000)')
    print('5초 대기시도')
    # 후기 개수를 확인 ==> 적정선까지 로드
    for cnt in driver.find_elements_by_id('count'):            # list라서 for문으로 돌린다.
        print( cnt.text )
    # 후기 채집( 작성자, 내용, 좋아요 갯 수 )
    for comment in driver.find_elements_by_tag_name('ytd-comment-thread-renderer'):
        print(comment.text)
        print('-'*20)
```
