```python
import numpy as np
```

```python
x = np.arange(0,12,0.001)
x
```

```python
y = np.sin(x)
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
plt.figure( figsize=(10,2))
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
plt.figure( figsize=(10,6))
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('값')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.grid()
plt.xlabel('time')
plt.ylabel('값')
plt.title('메롱')
plt.show


```

```python
plt.figure( figsize=(10,6))
plt.plot(x,y, label = 'sin')
plt.plot(x,np.cos(x), label = 'cos')
plt.legend()
plt.xlabel('time')
plt.ylabel('값')
plt.title('오잉')
plt.xlim(0,np.pi)
plt.ylim(-1.2,1.2)
plt.grid()
plt.show()
```

```python
t = np.arange(0,5,0.5)
plt.figure(figsize=(10,6))
plt.plot(t,t,'r--')
plt.plot(t,t**2,'bs')
plt.plot(t,t**3,'g^')

plt.show()
```
