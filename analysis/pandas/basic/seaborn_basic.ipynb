```python
# 씨본로드
import seaborn as sns
```

```python
# 뒷배경
sns.set_style('whitegrid')
```

```python
# 기본 데이터
# tips : 요일별 점심, 저녁, 흡연여부, 식사금액, 팁 금액, 성별, 인원수 )
tips = sns.load_dataset('tips')
tips
```

```python
# 데이터 표현
# 가격대가 어디에 몰려있는지
plt.figure(figsize=(10,6))
sns.boxplot( x=tips['total_bill'] )
plt.show()

# 그림을 보몀ㄴ 10~20 달러 사이에 몰려있다.
```

```python
tips.columns
```

```python
plt.figure(figsize=(10,6))
# x, y축에 대한 컬럼을 지정, 데이터를 별도로 대입
sns.boxplot( x='day', y='total_bill', data= tips )
plt.show()

# 영업을 4일을 하고 토요일에 매출이 제일 많다.
```

```python
plt.figure(figsize=(10,6))
# 위의 시각화 자료에 흡연 여부를 넣겠다.
sns.boxplot( x='day', y='total_bill', data= tips, hue='smoker', palette='Set3' )     # palette : 색추가
plt.show()

```

```python
plt.figure(figsize=(10,6))
# 무리형태로 -> 포인트 중심 시각화
sns.swarmplot( x='day', y='total_bill', data= tips, color='0.5' )     
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot( x='day', y='total_bill', data= tips, palette='Set3' )
sns.swarmplot( x='day', y='total_bill', data= tips, color='0.5' )     
plt.show()
```

### 산포도 + 평균선(x,y에 대한 1차함수)

```python

sns.set_style('darkgrid')
sns.lmplot( x='total_bill', y='tip', data=tips,size=10)
plt.show()
```

```python
# 흡연여부 추가
sns.lmplot( x='total_bill', y='tip', hue='smoker', data=tips,size=10)
plt.show()
```

```python
# matplotlib에 비해 코드량이나 작업량이 훨씬 간결하고 파워풀하다
sns.lmplot( x='total_bill', y='tip', hue='smoker', data=tips,size=10, palette='Set2') # palette (색) 추가, 
plt.show()
```

```python
flights = sns.load_dataset('flights')  # 내장된 데이터 flights(항공정보)
flights.shape
```

```python
flights.head
```

```python
flights['year'].unique(), len(flights['year'].unique())     # 년도별
```

```python
flights['month'].unique(), len(flights['month'].unique())   # 월별
```

```python
# 인덱스에 12개월, 컬럼에 년도, 데이터는 고객 수 
flights_piv = flights.pivot( 'month', 'year', 'passengers' )
flights_piv
```

```python
# 히트맵을 통해서 고객 수가 어느 방향( 년도, 월)으로 증가하고 있는지 기세를 확인
plt.figure(figsize=(10,6))
sns.heatmap( flights_piv )
plt.show()

# 밝은 쪽이 많은 부분
```

```python
plt.figure(figsize=(12,10))
# 데이터 세팅(그림에 숫자 세팅)
# annot: 숫자세팅
# fmt : 수치의 타입( d: 정수)
sns.heatmap( flights_piv, annot=True, fmt='d')
plt.show()
```

# ----------------------------------------------산포행렬----------------------------------------------------

```python
# 산포 행렬
iris = sns.load_dataset('iris')
iris.head(2)
```

```python
# 대각선 기준으로 대칭확인
sns.pairplot( iris) # hue: 꽃 종류별로 색
plt.show()
# iris 변수는 개라 4x4
```

```python
# 대각선 기준으로 대칭확인
sns.pairplot( iris, hue='species' ) # hue: 꽃 종류별로 색
plt.show()
# iris 변수는 개라 4x4
```

```python
sns.pairplot( iris, hue='species', x_vars=['sepal_width', 'sepal_length']
             ,y_vars=['petal_width', 'petal_length']
            )
plt.show()
```
