### 데이터 프레임 병합

- concat()
- merge()

```python
import numpy as np
import pandas as pd
```

```python
# 데이터를 준비
left = pd.DataFrame( {
    'key':list('1234'),
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
} )
right = pd.DataFrame( {
    'key': list('0123'),
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
} )
```

```python
right
```

```python
# 합치기 : 두 개의 df를 공통으로 존재하는 컬럼 key를 기준으로 merge() 수행
# 교집합의 맥락, key 컬럼에 공통된 값을 가진 데이터만 존재시킴
pd.merge( left, right, on = 'key')
```

```python
pd.merge(  right,left, on = 'key')
```

```python
# 교집합 + right
# right 기준으로 데이터를 다 맞춘다 -> left에만 존재하는 데이터는 Nan 처리
pd.merge( left, right, on = 'key', how='right')
```

```python
# 교집합 , inner 생략해도 동일
pd.merge( left, right, on = 'key', how='inner')
```

```python
# 합집합
pd.merge( left, right, on = 'key', how='outer')
```

## concat

```python
# 상하로 붙였다.
pd.concat( [left,right] ,sort=True )
```

```python
# 좌우로 붙인다.
pd.concat( [left,right] ,sort=True, axis=1 )   # 방향성 - axis : 0이면 상하 1이면 좌우
```
